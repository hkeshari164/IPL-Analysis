```python
# Finding top 15 players with most man of the match
ipl['player_of_match'].value_counts()[0:15]
```

```python
player_name = list(ipl['player_of_match'].value_counts()[0:15].keys())
```

```python
player_name
```

```python
#plotting a bar graph using matplotlib
plt.figure(figsize =(20,6))
plt.bar(player_name,list(ipl['player_of_match'].value_counts()[0:15]))
plt.title("Player with most man_of_the_match")
plt.show;
```

```python
#Counting of tied match, no result etc
frequency = ipl['result'].value_counts()
frequency
```

```python
# finding out the number of toss wins with respect to each team
toss_wins = ipl['toss_winner'].value_counts()
toss_wins
```

```python
# Finding the record where team won by batting first
batting_first = ipl[ipl['win_by_wickets'] == 0]
batting_first
```

```python
# Histogram
plt.figure(figsize = (15,6))
plt.hist(batting_first['win_by_runs'],bins =np.arange(0,150,10))
plt.title("distribution of runs")
plt.xlabel("runs")
plt.show()
```

Here we can see that teams winning by more than 100 runs are very less and more number of teams who won in the range of 10 to 20

```python
# Finding out the number of wins w.r.t each other afterbatting first
batting_first['winner'].value_counts()
```

```python
#Making a bar plot for top 5 teams for batting first 
plt.figure(figsize = (10,6))
plt.bar(batting_first['winner'].value_counts()[0:5].keys(), batting_first['winner'].value_counts()[0:5], color = ['red', 'yellow', 'green', 'blue', 'pink'])
plt.xlabel("Teams")
plt.xticks(rotation = 45)
plt.show()
```

```python
# Making a pie chart for showing distributions of top 5 teams
plt.figure(figsize = (6,6))
plt.pie(batting_first['winner'].value_counts()[0:5], labels = (batting_first['winner'].value_counts()[0:5].keys()), autopct= '%1.1f%%')
plt.show()
```

```python
# Extracting those records where a team won by batting second
batting_second = ipl[ipl['win_by_runs']==0]
batting_second
```

```python
# Histogram
plt.figure(figsize = (8,8))
plt.hist(batting_second['win_by_wickets'], bins = 10)
plt.show()
```

```python
#finding out the frequency of number of wins w.r.t each time after batting second
batting_second['winner'].value_counts()
```

```python
# making a bar plot for top 5 teams with most wins after batting second
plt.figure(figsize=(8,8))
plt.bar(batting_second['winner'].value_counts()[0:5].keys(), batting_second['winner'].value_counts()[0:5], color = ['red', 'blue', 'yellow','black', 'green'])
plt.xlabel('team')
plt.xticks(rotation = 35)
plt.show()
```

```python
# making a pie chart for distribution of most wins after batting second
plt.figure(figsize = (8,8))
plt.pie(batting_second['winner'].value_counts()[0:5], labels = (batting_second['winner'].value_counts()[0:5].keys()), autopct= '%1.1f%%')
plt.show()
```
